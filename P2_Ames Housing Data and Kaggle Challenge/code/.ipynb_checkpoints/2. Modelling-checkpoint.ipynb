# Project 2 - Ames Housing Data (Part 2 - Modelling)

## Imports

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='darkgrid')

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, ElasticNet, ElasticNetCV, Ridge, Lasso, RidgeCV, LassoCV
from sklearn.metrics import mean_squared_error, r2_score

# Display all rows and columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
train_df = pd.read_csv('../datasets/train_cleaned.csv')
train_df.head()
```

```python
train_df.info()
```

```python
train_df.shape
```

### One-hot encode categorical columns

```python
train_df['ms_subclass'] = train_df['ms_subclass'].apply(str)
```

The data type for 'ms_subclass' reverted back to an integer value. We shall convert it to string format again for one-hot encoding in the next step.

```python
train_df = pd.get_dummies(data=train_df,drop_first=True)
```

```python
train_df.shape
```

```python
train_df.head()
```

### Modelling

```python
#define X/y variables:
X = train_df.drop('saleprice',axis=1)
y = train_df['saleprice']

#train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

#### Baseline Model

```python
# Baseline Model
print('Baseline Model')
print(u'\u2500' * 30)
    
#metrics
baseline_rmse = np.sqrt(mean_squared_error(y_test,[np.mean(y_train)] * len(y_test)))
baseline_r2 = r2_score(y_test,[np.mean(y_train)] * len(y_test))
    
print(f"Baseline Model RMSE: ", baseline_rmse)
print(f"Baseline Model R2: ", baseline_r2)
```

#### Linear Regression Model

```python
#Linear Regression model
lr = LinearRegression()
run_model(lr,'Linear Regression Model', X_train, X_test, y_train, y_test)
```

#### Ridge Model

```python
#Ridge model

#Scale
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

ridge = Ridge()
run_model(ridge,'Ridge Model', X_train_scaled, X_test_scaled, y_train, y_test)
```

#### Lasso Model

```python
#Lasso model

#Scale
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

lasso = Lasso(max_iter=50000)
run_model(lasso,'Lasso Model', X_train_scaled, X_test_scaled, y_train, y_test)
```

#### Elastic Net Model

```python
#Elastic net model

#Scale
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

enet = ElasticNet()
run_model(enet,'Elastic Net Model', X_train_scaled, X_test_scaled, y_train, y_test)
```

From the above results from running baseline model, linear regression model, lasso model, ridge model and elastic net model on default parameters, proved to be much superior over the baseline model.

Comparing between the linear regression model, lasso model, ridge model and elastic net model, the first 3 had similar performances whereas the elastic net model fared slightly worse. For all 4 models mentioned, the RMSE and R2 scores between train and test datasets are similar, and the cross-validation score and the test R2 scores are comparable as well. This shows that these models do not have significant overfitting and underfitting.

We will then attempt to tune some of the hyperparameters to see if we can further improve our models' performances.

### Re-run models with hyperparameter tuning

#### Ridge Model

```python
#RidgeCV model

#Scale
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

ridgecv = RidgeCV(alphas=np.logspace(0, 5, 100), scoring='r2', cv=5)
run_model(ridgecv,'Ridge Model with Tuning Model', X_train_scaled, X_test_scaled, y_train, y_test)
```

#### Lasso Model

```python
#Lasso model

#Scale
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

lassocv = LassoCV(n_alphas=100,cv=5, max_iter=50000)
run_model(lassocv,'Lasso Model with Tuning Model', X_train_scaled, X_test_scaled, y_train, y_test)
```

#### Elastic Net Model

```python
#Elastic model with tuning

#Scale
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

#set hyperparameters
alphas = np.linspace(0.5, 50, 100)
l1_ratio = [0.1,0.3,0.5,0.7,0.95,0.99]

#instantiate model
enetcv = ElasticNetCV(alphas=alphas, l1_ratio=l1_ratio,cv=5)
run_model(enetcv,'Elastic Net with Tuning Model', X_train_scaled, X_test_scaled, y_train, y_test)
```

From the results of running our models with hyperparameter tuning, we see that the Lasso and Ridge Regression model have slightly worse performance and Elastic Net model having better performance.

It seems we have the flexibility to choose from a number of models according to the models' performance metrics. However, if we were to compare all of our models, pre and post hyperparameter tuning, our initial Linear Regression model, default Ridge Regression and default Lasso Regression model seemed to give us the best RMSE and R2 scores.

We ultimately decide to use the linear regression model as carrying out regularization did not seem to improve our model's performance. In addition, our linear regression model did not have a notable overfitting issue and thus there was no need to regularize our model, and we could also reduce complexity with just using the linear regression model.

### Visualization of predictions

```python
lr = LinearRegression()

lr.fit(X_train,y_train)
y_preds = lr.predict(X_test)
```

```python
g = sns.jointplot(y_test, y_preds, kind='scatter',height = 8 );
g.ax_joint.plot(y_test, y_test, linestyle = 'dashed', color = 'black');
g.set_axis_labels('Actual Price (USD)','Predicted Price (USD)', fontsize = 14);
```

From the figure above, we can see that our model is a good prediction of the actual sale price for prices below USD350,000. Above this sale price, our model appears to under-predict the actual sale price.

## Conclusions and Recommendations

The linear regression model we built was based on selective feature selection and engineering, and the final model to predict the sale prices of the properties achieved an accuracy score of 86.3%, based on the cross-validation score we obtained for the linear regression model. The model also achieved a root mean squared error(RMSE) of 24338 and a R2 score of 0.9, representing a model that has a rather low error rate.

This would give property traders in Ames, the confidence to purchase a property if they deem it sufficiently undervalued, as well as prospective homebuyers to purchase a residential property if the model deems that the property is fairly valued.

However, our model does come with shortcomings, in that the accuracy of the model starts dropping when the sale prices exceed USD350,000. In such scenarioes, users would be advised to utilize the model for comparison's sake instead and not to over-rely on the absolute values churned by the model. To improve on the shortcomings, we could possibly further refine the model upon collecting more data of houses with sale prices above USD 350,000.
